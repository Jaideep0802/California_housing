```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()
```

```python
print(housing)
```

```python
housing_dataframe = pd.DataFrame(housing.data, columns=housing.feature_names)
```

```python
print(housing_dataframe)
```

```python
housing_dataframe['price'] = housing.target
```

```python
housing_dataframe.head()
```

```python
housing_dataframe.shape
```

```python
housing_dataframe.isnull().sum()
```

```python
housing_dataframe.describe()
```

```python
correlation = housing_dataframe.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
X = housing_dataframe.drop(['price'],axis=1)
Y = housing_dataframe['price']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_test.shape,X_train.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data = model.predict(X_train)
```

```python
print(training_data)
```

```python
score_r2 = metrics.r2_score(Y_train,training_data)
```

```python
score_mae = metrics.mean_absolute_error(Y_train,training_data)
```

```python
print('R2 Score:',score_r2)
print('MAE Score:',score_mae)
```

```python
test_data = model.predict(X_test)
```

```python
print(test_data)
```

```python
score_r2_2 = metrics.r2_score(Y_test,test_data)
```

```python
score_mae_2 = metrics.mean_absolute_error(Y_test,test_data)
```

```python
print('R2 Score:',score_r2_2)
print('MAE Score:',score_mae_2)
```

```python
plt.scatter(Y_train,training_data)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
